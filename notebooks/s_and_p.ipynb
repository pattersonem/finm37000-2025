## S&P Futures: ES

These Mini contracts took over the non-mini standard `SP` future by providing a smaller contract
size (\\$50 x S&P index vs. \\$250 x S&P index) and better electronic execution (`SP` remained
floor traded when `ES` was trading online).

N.B., micro contracts are also available (\\$5 x S&P index)
but not nearly as popular with open interest about 10% of the mini (so 1% of the dollar value given
tha contract size difference) and trade volume sometimes on par with the mini but 10% of the value.

Key features of `ES`:
* Value \\$50 x S&P index.
* Tick size 0.25 index points, i.e., \\$12.50.
* Calendar spreads tick in 0.05 index points.
* Volume/Open interest highly concentrated in the front month.
* Financially settled.

## ES Options

By `ES` options, I mean all CME options on the underlying `ES` future. There are actually many different symbols
associated with these options. Most follow a tick size convention 0.25 index points for premium above 10 index points
and 0.05 for premium at or below 10 index points.

* `ES` options
    * American exercise
    * quarterly expirations, 9:30am ET on third friday.
* `EW` end-of-month (EOM) options
    * European exercise
    * End-of-month expiration, last business day, 4pm ET.
* Weekly/daily options
    * European exercise
    * Do not exist when they would coincide with the EOM `EW` option.
    * Daily expirations at 4pm ET available in week-of-month/day-of-week groups
        * Monday: `E1A`, `E2A`, `E3A`, `E4A`, `E5A`
        * Tuesday: `E1B`, `E2B`, `E3B`, `E4B`, `E5B`
        * Wednesday: `E1C`, `E2C`, `E3C`, `E4C`, `E5C`
        * Thursday: `E1D`, `E2D`, `E3D`, `E4D`, `E5D`
        * Friday: `EW1`, `EW2`, `EW3`, `EW4`, `EW5`

See daily volume and open interest numbers here:
https://www.cmegroup.com/markets/equities/us-index/sp-index-futures-and-options.html#tab_LOV9XAi=options

or here:
https://www.cmegroup.com/market-data/browse-data/equity-volume.html?redirect=/market-data/volume-open-interest/equity-volume.html

Usual pattern of trading activity is
* Expiring daily contract (0DTE) has the most volume (hundreds of thousands) but ending with minimal open interest.
* Next daily contract has significant volume and open interest has been building for a few weeks into the millions.
* Quarterly `ES` has significant volume (hundreds of thousands), but not as much as the most active dailies, and open interest often
comparable to biggest daily contract or more (millions).
* EOM `EW` has significant volume and open interest comparable to the quarterly.

Total volume across all types can easily reach a million with open interest several million.

## SPX Options

`SPX` and `SPXW` options are Cboe products https://www.cboe.com/tradable-products/sp-500/spx-options
* Cash settled.
* European expiration.
* `SPX` are settled based on opening prices for constituent stocks in the index on the third Friday.
* `SPXW` provide M-F expirations, 3rd Friday, and EOM afternoon settlement based on index prices.
* Contract size is \\$100 x S&P Index (i.e., 2x `ES` options).
* Minimum tick size is \\$0.05 for premium below 3.00 index points, \\$0.10 for premium above 3.00.
* Also have mini `XSP` (1/10th) and nano contracts with smaller size and smaller trade volumes.

## VIX - the index

How do you summarize the implied volatility across all strikes in market?

$$
\sigma^2 = \frac{2}{T} \sum_{i} \frac{\Delta K_i}{K_i^2}e^{RT} Q(K_i) - \frac{1}{T}\left[ \frac{F}{K_0} - 1 \right]^2
$$

VIX index is $ 100 \times \sigma$.

Calculation details:
* $Q(K_i)$ is the OTM bid-ask mid price for strike $K_i$.
* $F$ is option-implied forward price: $F = K - e^{RT}(C - P)$.
* $K_0$ is ATM strike at or below $F$.
* $\Delta K_i$ is the width of the strike grid around $K_i$
* The set of $i$ is determined by which strikes have bid-ask markets
    * Exclude strikes with missing/zero bid.
    * Expanding out from ATM $K_0$, stop inclusion once two consecutive strikes are excluded.

Key steps to calculate a 30-day volatility:
1. Determine `SPX` or Friday `SPXW` expirations bracketing 30-days to expiration, near and next term expirations.
2. Determine a near and next term interest rate using cubic splines on Treasury rates.
3. Calculate $\sigma^2$ for both near and next term expirations.
4. Weight the two $\sigma^2$ values to get a constant maturity 30-day implied volatility estimate.

Details of the current version of the methodology at the Cboe:
 https://cdn.cboe.com/resources/indices/Volatility_Index_Methodology_Cboe_Volatility_Index.pdf

## VIX - the future

The interest in the S&P index and its volatility led to the introduction of futures on the VIX index
in 2004
traded at the Cboe Futures Exchange (CFE).

The methodology to determine settlement prices on (typically) Wednesday mornings uses
a _single_ expiration of `SPX` or `SPXW` options based on opening trades
for a designated range of strikes rather than bid-ask mid prices.

Futures trade:
* Take a view or a hedge on option-implied vol
* Not necessary to manage option trading complexity.

Basis trade:
* `SPX` options are the components of the index.
* Cost-of-carry less important than other futures (what are your storage costs for having an options position?).
* More focused on the option market vs. the VIX index methodology.

Trading in VIX futures drives additional trading in the underlying SPX option components,
contributing to additional trading volume and open interest in SPX options.

## VIX Options

Options on the VIX index were introduced in 2006 and financially settled, not to be confused with UX options,
a relatively tiny market that settles into VIX futures.

The VIX option market is active with many features that make it quite different than other options
markets: call skew is much more active than put skew.

The same lessons we learned about trading options on futures can be applied to trading options on
the VIX index.
